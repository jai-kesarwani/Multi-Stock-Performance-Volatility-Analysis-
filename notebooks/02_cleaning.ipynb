```python
import pandas as pd
```

```python
# Convert date
df['Date']=pd.to_datetime(df['Date'])
print(df['Date'].dtype)
```

```python
#  checking null values
df.isnull().sum()
```

```python
# Checking duplicate values
df.duplicated().sum()
```

```python
# Chceking if there is any negative price of stock 
(df[['Open','High','Low','Close']] < 0).sum()
```

```python
#  Validating volume 
(df['Volume'] < 0).sum()
```

```python
# Checking OHLC Consistency 
df = df[
    (df['High']>=df['Open']) &
    (df['High']>=df['Close']) &
    (df['Low']<=df['Open']) &
    (df['Low']<=df['Close'])
]
```

```python
# adding date feature

df['Year'] = df['Date'].dt.year

df['Month'] = df['Date'].dt.month

df['Quarter'] = df['Date'].dt.quarter

df['Day'] = df['Date'].dt.day
```

```python
df.to_csv(
    "../data/cleaned_stock_data.csv",
    index=False
)

print("Saved successfully")
```
