```python
import yfinance as yf
import pandas as pd

stocks = [
    'AAPL',
    'MSFT',
    'GOOGL',
    'AMZN',
    'TSLA',
    'NVDA'
]

all_data = []

for stock in stocks:

    data = yf.download(
        stock,
        start='2020-01-01',
        end='2025-01-01'
    )

    # Remove second level column names
    data.columns = data.columns.droplevel(1)

    data.reset_index(inplace=True)

    data['Stock'] = stock

    all_data.append(data)

final_df = pd.concat(
    all_data,
    ignore_index=True
)

print(final_df.columns)

final_df.to_csv(
    "raw_stock_data.csv",
    index=False
)
```
