```python
import pandas as pd

df = pd.read_csv(
    "../data/cleaned_stock_data.csv"
)

df['Date'] = pd.to_datetime(
    df['Date']
)

df.head()
```

```python
# Sorting the data

df = df.sort_values(
    ['Stock','Date']
)

df.head()
```

```python
# DAILY RETURN measures % change in price.
# RETURN = (Pt-(Pt-1)/(Pt-1))

df['Daily_Return']=(df.groupby('Stock')['Close']).pct_change()
```

```python
df[['Date','Stock','Close','Daily_Return']].head()
```

```python
# Volatility => higher volatitlity -> higher risk
# 30 day rolling risk

df['Volatility']=(df.groupby('Stock')['Daily_Return'].transform(lambda x:x.rolling(30).std()))
```

```python
df[['Stock','Volatility']].tail()
```

```python
# Moving average (50 days MA)

df['MA50'] = (df.groupby('Stock')['Close'].transform(lambda x:x.rolling(50).mean()))
```

```python
# 200 days MA

df['MA200'] = (df.groupby('Stock')['Close'].transform(lambda x:x.rolling(200).mean()))
```

```python
# CUMMULATIVE RETURN -> Shows total growth over time.

df['Cumulative_Return'] = (1+df['Daily_Return']).groupby(df['Stock']).cumprod()
```

```python
df[['Stock','Cumulative_Return']].tail()
```

```python
#  Rolling Average Volume 30 days

df['Avg_Volume_30'] = (df.groupby('Stock')['Volume'].transform(lambda x:x.rolling(30).mean()))
```

```python
# Annualized Volatility

import numpy as np

df['Annualized_Volatility'] = (df['Volatility']*np.sqrt(252))
```

```python
# Sharpe Ratio 

risk_free_rate = 0.02

mean_return = (df.groupby('Stock')['Daily_Return']
                .transform('mean'))
                
df['Sharpe_Ratio'] = (mean_return-risk_free_rate/252)/df['Volatility']
```

```python
# Checking nulls

df.isnull().sum()
```

```python
df.dropna(
    inplace=True
)
```

```python
df.isnull().sum()
```
